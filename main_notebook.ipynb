```python
def parse_extras(row):
    row = str(row)
    pairs = row.split('<->')
    values = {}
    for pair in pairs:
        if pair.strip() != '' and pair.strip() != 'nan' and pair.strip() != "0":
            col = pair
            val = pair
        else:
            col = "trash"
            val = np.nan
        values[col] = val
    return pd.Series(values)

extras = data['extras_types'].apply(parse_extras).apply(pd.Series)
extras = extras.drop("trash", axis=1)
```
